```python
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse, unquote
from IPython.display import Image, SVG, display, HTML
import requests
import os
import mimetypes

url = "https://en.wikipedia.org/wiki/Wikipedia:Featured_pictures"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36",
    # Referer helps with some servers that block hotlinking
    "Referer": url
}

resp = requests.get(url, headers=headers, timeout=15)
resp.raise_for_status()

html_content = BeautifulSoup(resp.content, "html.parser")
img_tags = html_content.find_all("img")

folder = "images"
os.makedirs(folder, exist_ok=True)

def safe_filename_from_url(img_url, content_type=None):

    path = urlparse(img_url).path
    name = os.path.basename(path) or "image"
    name = unquote(name)

    name = name.split("?")[0]

    if "." not in name:
        if content_type:
            ext = mimetypes.guess_extension(content_type.split(";")[0].strip())
            if ext:
                name = name + ext
    return name

for img in img_tags:
    src = img.get("src") or img.get("data-src") or img.get("data-image-name")
    if not src:
        continue


    if src.startswith("data:"):
        print("Skipping data URI")
        continue

    img_url = urljoin(url, src)
    print("Requesting:", img_url)

    try:
        r = requests.get(img_url, headers=headers, stream=True, timeout=15)
    except Exception as e:
        print("Request error:", e)
        continue

    print("  Status:", r.status_code)
    ctype = r.headers.get("Content-Type", "")
    print("  Content-Type:", ctype)

    if r.status_code != 200:
        print("  Not 200 â€” skipping")
        continue


    if not ("image" in ctype or "svg" in ctype):
        print("  Response doesn't look like an image (maybe HTML). Skipping.")

        continue

    filename = safe_filename_from_url(img_url, content_type=ctype)
    out_path = os.path.join(folder, filename)

    try:
        with open(out_path, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

        size = os.path.getsize(out_path)
        print(f"  Saved: {out_path}  ({size} bytes)")


        if "svg" in ctype:
            try:
                display(SVG(out_path))
            except Exception as e:

                display(HTML(f'<img src="{out_path}" alt="{filename}" />'))
        else:
            display(Image(filename=out_path))

    except Exception as e:
        print(f"  Failed to write/display: {e}")
```
